```python
import requests
from tmdbv3api import TMDb, Movie
from urllib3.exceptions import MaxRetryError

# Initialize TMDb and Movie objects
tmdb = TMDb()
tmdb_movie = Movie()

def get_genre(x):
    genres = []
    try:
        result = tmdb_movie.search(x)
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id, tmdb.api_key), verify=True)
        if response.status_code == 200:
            data = response.json()
            genres = [genre['name'] for genre in data.get('genres', [])]
    except MaxRetryError as e:
        print(f"Max retries exceeded. Error: {e}")
    except requests.exceptions.SSLError as e:
        print(f"SSL Error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")
    return genres

# Assuming you have a DataFrame called 'df' and a column 'Title'
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```
